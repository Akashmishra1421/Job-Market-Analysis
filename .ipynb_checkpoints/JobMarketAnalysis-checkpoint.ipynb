```python
import pandas as pd
```

```python
df = pd.read_csv('clean_jobs.csv')
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['work_type','employment_type'])
```

```python
new_df = df.drop(columns=['work_type','employment_type'])
```

```python
new_df.info()
```

```python
new_df.fillna(value='unknown')
```

```python
new_df = new_df.fillna(value='unknown')
```

```python
new_df.info()
```

```python
new_df['date_posted'] = pd.to_datetime(new_df['date_posted'],format= 'ISO8601',errors='coerce')
```

```python
new_df
```

```python
new_df['month_wise']= new_df['date_posted'].dt.month
```

```python
new_df.info()
```

```python
new_df['month_wise'] = new_df['month_wise'].fillna(0)
```

```python
new_df['month_wise' ]= new_df['month_wise'].astype(int)
```

```python
new_df
```

```python
import matplotlib.pyplot as plt
```

```python
y = df.groupby(new_df['month_wise'])['id'].count()
x = y.index
```

```python
plt.xlabel('Month')
plt.ylabel('Job Count')
plt.bar(x,y)
plt.show()
```

```python
month_names = {0: 'Unknown', 1: 'January', 
               2: 'February', 3: 'March', 4: 'April',5:'May',6:'June'}
```

```python
x = x.map(month_names)
```

```python
plt.xlabel('Month')
plt.ylabel('Job Count')
plt.title('Month Wise Job opening Trends')
plt.bar(x,y)
plt.show()
```

```python
new_df['description'].head()
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
description = ' '.join(new_df['description'].str.lower()).split()
```

```python
words = [word for word in description if word not in stop_words]
```

```python
from collections import Counter
```

```python
Counter(words).most_common(20)
```

```python
extra_stopwords = ['*', 'de', 'work', 'team', 
'working', 'across', 'years', 'role', 
'including', 'help', 'ability', 'support','new', 'using', 'us', 'people', 
'benefits','show', '&', 'job', 'related', 
'opportunity','e']
```

```python
stop_words.update(extra_stopwords)
```

```python
words = [word for word in description if word not in stop_words]
```

```python
Counter(words).most_common(20)
```

```python
word_freq = Counter(words).most_common(20)
```

```python
from wordcloud import WordCloud
```

```python
Word_Cloud = dict(word_freq)
```

```python
wc = WordCloud(width=800, height=400, background_color='white')
wc.generate_from_frequencies(Word_Cloud)
```

```python
plt.title('Most In-Demand Skills in Data Analyst Jobs')
plt.imshow(wc)
plt.axis('off')
plt.show()
```
